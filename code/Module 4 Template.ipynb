* Cancer hallmark focus: The focus for this project is on two primary hallmarks, of resisting apoptosis and genomic instability.
* Overview of hallmark: 

Resist apoptosis: mutations in certain genes cause damaged cells to multiply, instead of undergoing apoptosis. For example, a mutation in the TP53 gene produces a faulty p53 transcription factor. Normally, p53 helps to destroy a damaged cell, but the mutant p53 continues to allow for cell growth. This genetic mutation is seen in many different cancer types, being present in 50% of all human tumor growths (Hamzehloie et al., 2012). 

Genome instability: this characteristic of cancer causes cancer cells to divide into different mutated “subclones”, causing heterogeneity and diversity in cancer cells within the same tumor. Additionally, the same cancer type can develop into tumors that look vastly different between different patients. LGGs are a heterogeneous set of gliomas affected by this genome instability cancer hallmark (Cao et al., 2022).

The relationship between TP53 and BIRC5:
TP53: this gene is known as the "guardian of the genome" for its integral role in regulating cell death. TP53 encodes for p53, which is a transcription factor that signals for the destruction of damaged cells. In LGG, mutant TP53 causes damaged cells to resist cell death and continue to grow uncontrollably, additionally contributing to the genome instability hallmark.

BIRC5: this gene is also known as survivin, and has been shown to increase cell proliferation by inhibiting apoptosis (Jiang et al., 2022). Higher expression of BIRC5 has been shown to be positively correlated to more adverse manifestations of LGG. TP53 usually represses BIRC5; however, mutant TP53 lifts the control and may cause a surge in BIRC5, preventing the damaged cell from dying.

The relationship between ATRX and MCL1:
ATRX: this gene is involved in the stabilization chromatin and cell cycle regulation. Without ATRX, DNA damage and replicative stress has been shown to occur, leading to genome instability.

MCL1: this gene is found to be frequently amplified in all human cancers, including LGG (Zhang et al., 2023). With loss of ATRX, there is expected to be higher MCL1 levels in order to combat the genomic instability caused by the mutant ATRX.

The relationship between MGMT and BAX:
MGMT: MGMT normally protects against damage alkylating DNA damage. However, in LGG, this gene can be silenced and causes cells to accumulate mutations and further genomic instability. 

BAX: this gene normally responds to DNA damage and causes apoptosis. However, in LGG, lower expression of this gene can allow cells with damaged DNA that went unchecked by MGMT to persist and survive.

* Prevalence & incidence

Low-grade gliomas (LGGs) have an incidence rate of around 1.12 per 100,000 people (Cao et al., 2023). When examining sex differences, incidence rates were lower in females than males. Different subtypes of LGGs also had different incidence rates, with the highest incidence rate being astrocytomas (0.43 per 100,000 people). Incidence rates peak in individuals aged 30-40 (Forst et al., 2014).

* Risk factors (genetic, lifestyle) & Societal determinants
Exact risk factors of LGGs are not well understood. However, previous exposure to ionizing radiation or radiation therapies can increase the risk of developing LGGs (Karsonovich et al., 2025). Additionally, LGGs are more common in older individuals, and is more common in males (Cao et al., 2023).

From a genetic standpoint, having a specific allele of CCDC26 on chromosome 8 has also been identified as a risk factor, with this allele seen in 40% of patients with certain gliomas, while only 8% of the general population has this allele (Toader et al., 2023). Additionally, mutations in isocitrate dehydrogenase (IDH) 1 and 2 are present in more than 80% of LGG tumors (Bready & Placantonakis, 2019).

* Standard of care treatments (& reimbursement) 

Surgery to remove the tumor is the primary treatment method for LGG. For grade 1 gliomas, surgery may commpletely removing the tumor and cure the disease. However, if the cancerous growth is not completely removed by surgery, other treatment options include targeted therapies, chemotherapy, and radiation therapy. Targeted therapies is becoming a top treatment option, as medicines can specifically target the cancer cells and shrink the tumor without as many side effects as chemotherapy or radiation therapy (Low-grade glioma treatment).

* Biological mechanisms (anatomy, organ physiology, cell & molecular physiology)

Gliomas are a subset of central nervous system (CNS) tumors that affect glial cells. Glial cells are the most abundant cells in the CNS, and provide essential functions such as forming myelin, and supporting neurons in the nervous system (Ludwig & Das, 2023). Low-grade gliomas (LGG) include grade 1 and grade 2 gliomas, which are relatively slow-growing and less aggressive (Toader et al., 2023).

Genetic mutations in tumor suppressor protein 53 (p53) and epidermal growth factor receptor (EGFR) contribute to LGG growth. The mutated p53 does not function properly to destroy cells with faulty DNA, and amplificatin of EGFR causes cells to divide uncontrollably (Karsonovich et al., 2025). Additionally, mutations in IDH1 and 2 increase tumorigenesis (Schaff et al., 2024). The result of these mutations is slow growth of "undifferentiated, proliferating stem-like tumor cells" (Bready & Placantonakis, 2019).

The dataset was aquired using next generation sequencing methods to allow millions of cDNA fragments to be sequenced simultaneously. Then, the raw sequences are mapped back the the reference human genome to determine which specific genes the RNA fragments came from and allows for their number to be quantified.

Because the raw counts can be biased by gene length or the total depth of the sequencing run, the data is normalized. Reads per kilobase (RPK) adjusts for the fact that longer genes naturally produce more fragments, transcipts per million (TMP) further normalizes the number of reads of a sample, making it possible to compare different cancer samples directly, and log-transformation of the TPM values stabilizes the variance in the data, making it easier to perform statistical comparisons between cancer types.

This dataset was collected from The Cancer Genome Atlas RNA-seq dataset, re-processed by Rahman et al

To analyze the molecular landscape of Brain Lower Grade Glioma (LGG), the code utilizes the pandas library to perform high-precision filtering on the TCGA RNA-seq dataset. We first filtered by cancer type, targeting the "cancer_type" column and isolating the "LGG" label. Once these specific samples are extracted, the code subsets the expression data to only include the genes related to the hallmarks of Genome Instability and Resisting Apoptosis, as described in the hallmarks background section. This ensures that the subsequent statistical analysis is both computationally efficient and biologically relevant.

Following the isolation of these hallmark signatures, the code prepares the data for quantitative comparison by transposing the matrix data so that individual genes serve as variables across the patient cohort. The code then generates descriptive statistics (such as mean, variance, and standard deviation) to evaluate the distribution of expression levels across the LGG samples, ensuring the data is suitable for linear analysis. This statistical grounding allows for the calculation of correlation coefficients for the three specific gene pairs, determining the strength and direction of the relationships between genome integrity and apoptotic resistance. Finally, the analysis produces linear regression plots that visually map these interactions, such as the coupling between  TP53 and  BIRC5, providing insight into how glioma cells may decouple their internal damage sensors from their programmed cell death machinery.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. LOAD DATA
# ------------------------------------------------------------------
data = pd.read_csv('../Data/TRAINING_SET_GSE62944_subsample_log2TPM.csv', index_col=0, header=0)
metadata_df = pd.read_csv('../Data/TRAINING_SET_GSE62944_metadata.csv', index_col=0, header=0)

# 2. FILTER FOR LGG SAMPLES
# ------------------------------------------------------------------
lgg_samples = metadata_df[metadata_df['cancer_type'] == 'LGG'].index
lgg_data = data[lgg_samples]

# 3. DEFINE THREE SPECIFIC PAIRS (Genome Instability vs. Apoptosis)
# ------------------------------------------------------------------
pairs = [
    ('TP53', 'BIRC5'), # The "Master Regulator" link
    ('ATRX', 'MCL1'),  # The "Chromatin/Survival" link
    ('MGMT', 'BAX')    # The "Repair/Death" link
]

target_genes = list(set([gene for pair in pairs for gene in pair]))
available_genes = [g for g in target_genes if g in lgg_data.index]
missing = set(target_genes) - set(available_genes)

if missing:
    print(f"CRITICAL EVALUATION: {missing} not in dataset. Skipping missing pairs.\n")

lgg_subset = lgg_data.loc[available_genes].T

# 4. SUMMARY STATISTICS
# ------------------------------------------------------------------
print("--- DESCRIPTIVE STATISTICS FOR HALLMARK GENES (log2TPM) ---")
# .describe() provides count, mean, std, min, 25%, 50%, 75%, and max
stats = lgg_subset.describe().T
# Adding variance specifically as it is useful for expression spread
stats['variance'] = lgg_subset.var()
print(stats[['mean', '50%', 'std', 'variance', 'min', 'max']])

print("\n--- PAIRWISE CORRELATION SUMMARY ---")
correlation_results = []
for gi_gene, ra_gene in pairs:
    if gi_gene in available_genes and ra_gene in available_genes:
        r_val = lgg_subset[gi_gene].corr(lgg_subset[ra_gene])
        correlation_results.append({'Pair': f"{gi_gene} vs {ra_gene}", 'Pearson_r': r_val})
        print(f"Correlation between {gi_gene} and {ra_gene}: {r_val:.4f}")

# 5. VISUALIZE THE THREE PAIRS
# ------------------------------------------------------------------
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, (gi_gene, ra_gene) in enumerate(pairs):
    if gi_gene in available_genes and ra_gene in available_genes:
        corr_val = lgg_subset[gi_gene].corr(lgg_subset[ra_gene])
        
        sns.regplot(ax=axes[i], data=lgg_subset, x=gi_gene, y=ra_gene, 
                    scatter_kws={'alpha':0.4, 'color':'navy'}, line_kws={'color':'crimson'})
        
        axes[i].set_title(f"{gi_gene} vs {ra_gene}\nPearson r = {corr_val:.2f}")
        axes[i].set_xlabel(f"{gi_gene} (Instability Marker)")
        axes[i].set_ylabel(f"{ra_gene} (Apoptosis Marker)")
    else:
        axes[i].text(0.5, 0.5, 'Gene Missing', ha='center', fontsize=12, color='red')

plt.tight_layout()
plt.suptitle("Hallmark Interactions in Brain Lower Grade Glioma (LGG)", y=1.05, fontsize=16)
plt.show()
```
